```python
import os
import shutil
from tqdm.notebook import tqdm
def underSamp(source_dir, dest_dir, num_copies):

    os.makedirs(dest_dir, exist_ok=True)


    for file in tqdm(os.listdir(source_dir)):
        if file.endswith(".jpg"):  # Ensure it's an image file
            src_path = os.path.join(source_dir, file)

            for i in range(int(num_copies)):
                new_file_name = f"{os.path.splitext(file)[0]}_copy{i}{os.path.splitext(file)[1]}"
                dest_path = os.path.join(dest_dir, new_file_name)
                shutil.copy(src_path, dest_path)

```

```python

def makeLabel(img_name,top5,top5conf):
  mapper = {0: 'Broad', 1: 'Broad+Pulse', 2: 'NAN', 3: 'Narrow', 4: 'Narrow+Broad', 5: 'Narrow+Pulse', 6: 'Pulse', 7: 'Uncertain', 8: 'Unknown+Pulse', 9: 'Unlabeled'}
  a = top5[0]
  b = top5[1]
  map_a = mapper[a]
  map_b = mapper[b]
  if top5conf[1]>0.1 and a!=2 and b!=2:
    if ((a in [7,9]) or (b in [7,9])) :
      if(a in [7,9]):
        return map_b
      else:
        return map_a
    elif (map_a in map_b )or (map_b in map_a) :
      if(len(map_a) > len(map_b)):
        return map_a
      else:
        return map_b
    elif (a==8 or b ==8):
      if a==8:
        return map_a.replace("Unknown",map_b)
      else:
        return map_b.replace("Unknown",map_a)
    else:
      return map_a
  else:
    return map_a
```

```python
from tqdm.notebook import tqdm
path = 'test'
img_name_list = os.listdir(path)

data_x = pd.DataFrame({
    'img_name': img_name_list,
    'top5': '',
    'top5conf': '',
})

n = len(img_name_list)

bz = 64
top5 = []
top5conf = []
for i in tqdm(range(0,n,bz)):
    if i+bz > n:
      bz = n-i
    images = [os.path.join(path,name) for name in img_name_list[i:i+bz]]
    preds = model.predict(images, verbose=False)
    for p in preds:
      top5.append(p.probs.top5)
      top5conf.append([round(px, 2) for px in p.probs.top5conf.cpu().tolist()])
    # print(x)
    # break

data_x['top5'] = top5
data_x['top5conf'] = top5conf

data_x.to_csv(f'pred_{path}.csv', index=False)
data_x
```
